```python
import numpy as np
import pandas as pd
import json
import os

import udpipe_parser
import nltk
# import stanfordnlp
import stanza # https://github.com/stanfordnlp/stanza
```

```python
with open("train.json") as infile:
    opener_en_train = json.load(infile)
```

```python
opener_en_train
```

```python
reqdKeys = {'Source', 'Target', 'Polar_expression'}
```

```python
len(opener_en_train)
```

```python
# Getting data for passing through BERT
cnt = 0
myfile = open('/media/anuradha/Data/Project/Sentiment analysis/data/target.txt', 'w')
data = []
for idx in range(0, len(opener_en_train)):
    if len(opener_en_train[idx]['opinions']) > 0 and (opener_en_train[idx]['opinions'][0].keys()) >= reqdKeys:
        txt = opener_en_train[idx]['text']
        print(idx)
        if len(opener_en_train[idx]['opinions'][0]['Source'][0]) > 0 and len(opener_en_train[idx]['opinions'][0]['Target'][0]) > 0 and len(opener_en_train[idx]['opinions'][0]['Polar_expression'][0]) > 0:
            data.append([txt, opener_en_train[idx]['opinions'][0]['Source'][0], opener_en_train[idx]['opinions'][0]['Target'][0], opener_en_train[idx]['opinions'][0]['Polar_expression'][0]])
#     df.append(arr)
myfile.close()
```

```python
len(data)
```

```python
data
```

```python
dfBERT = pd.DataFrame(data, columns = ['Text', 'Source', 'Target', 'Polar_expression'])
```

```python
dfBERT.to_csv('enBERT.csv')
```
